```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("train.csv")
```

```python
def wrangling(df):
    
    cat=[i for i in df.columns if df[i].dtype=="O"]
    notCat=[i for i in df.columns if(i not in cat)]
    
    
#     replacing all nan values with missing in HomePalnet
    df.HomePlanet.fillna("Missing_Home_planet",inplace=True)
#     now since homePlanet is an cetegorical type field we'll use dummies in here
    homePlanet=pd.get_dummies(df.HomePlanet)
    df=pd.concat([df,homePlanet],axis=1)
    df.drop(["HomePlanet","Missing_Home_planet"],axis=1,inplace=True)
#     work of homePlanet column is over now--dropping missing planet cause might be in testing dataset there is no missing columns
    
    
#     now procesing cryosleep column
    df.CryoSleep.fillna("Missing_Cryo_Sleep",inplace=True)
    cryoSleep=pd.get_dummies(df.CryoSleep,prefix="CryoSleep")
    df=pd.concat([df,cryoSleep],axis=1)
    df.drop(["CryoSleep","CryoSleep_Missing_Cryo_Sleep"],axis=1,inplace=True)
#    pre-procesiing of cryosleep column is done


#     now we'll deal with cabin column and its adverseries
    #df["Cabin_deck"]=df.Cabin.apply(lambda x: "Missing_deck" if str(x)=='nan' else list(x)[0])
    df["Cabin_deck"]=df.Cabin.apply(lambda x: "Missing_deck" if( len(list(str(x).split("/")))!=3 ) else list(str(x).split("/"))[0]   )
    #df["Cabin_side"]=df.Cabin.apply(lambda x: "Missing_side" if str(x)=='nan' else list(x)[4])
    df["Cabin_side"]=df.Cabin.apply(lambda x: "Missing_side" if( len(list(str(x).split("/")))!=3 ) else list(str(x).split("/"))[2]   )
    df.drop(["Cabin"],axis=1,inplace=True)
    dummy_cabin_deck=pd.get_dummies(df.Cabin_deck,prefix="Cabin_deck")
    dummy_cabin_side=pd.get_dummies(df.Cabin_side,prefix="Cabin_side")
    df=pd.concat([df,dummy_cabin_deck,dummy_cabin_side],axis=1)
    df.drop(["Cabin_deck","Cabin_side","Cabin_deck_Missing_deck","Cabin_side_Missing_side"],axis=1,inplace=True)
#     pre-processing for cabin is done

#     Destination pre processing
    df.Destination.fillna("Missing_Destination",inplace=True)
    destinantion=pd.get_dummies(df.Destination,prefix="Destination")
    df=pd.concat([df,destinantion],axis=1)
    df.drop(["Destination","Destination_Missing_Destination"],axis=1,inplace=True)
#     finally destination work is too done


#     working on VIP feature
    df.VIP.fillna("Missing_VIP",inplace=True)
    vip=pd.get_dummies(df.VIP,prefix="VIP")
    df=pd.concat([df,vip],axis=1)
    df.drop(["VIP","VIP_Missing_VIP"],axis=1,inplace=True)
    
    
    
#     now lets drop Name column cause it's of no fucking use
    df.drop(['Name'],axis=1,inplace=True)
    
#     now let's deal with numerical columns
    
    df.Age.fillna(df.Age.mean(),inplace=True)
    
    df.RoomService.fillna(0,inplace=True)
    df.FoodCourt.fillna(0,inplace=True)
    df.ShoppingMall.fillna(0,inplace=True)
    df.Spa.fillna(0,inplace=True)
    df.VRDeck.fillna(0,inplace=True)
    
    
    return df
    
```

```python
edt=wrangling(df)
```

```python
from sklearn.preprocessing import LabelEncoder
enc=LabelEncoder()
edt.Transported=enc.fit_transform(edt.Transported)
edt.head()
```

```python
df.head()
```

```python
x=edt.drop(["PassengerId","Transported"],axis=1)
y=edt.Transported
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train,y_train)
model.score(x_test,y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
nes=[]
acc=[]
for i in range(1,100):
    model2=RandomForestClassifier(n_estimators=i)
    model2.fit(x_train,y_train)
    nes.append(i)
    acc.append(model2.score(x_test,y_test))
print(nes)
print(acc)
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV
```

```python
pipelines={
    'rf':make_pipeline(StandardScaler(),RandomForestClassifier(random_state=1234)),
    'gb':make_pipeline(StandardScaler(),GradientBoostingClassifier(random_state=1234))
    
}
```

```python
RandomForestClassifier().get_params()
```

```python
grid={
    'rf':{
      'randomforestclassifier__n_estimators':[100,200,300]  
    },
    'gb':{
        'gradientboostingclassifier__n_estimators':[100,200,300]
    }
}
```

```python
fit_model={}
for algo,pipeline in pipelines.items():
    print(f"training {algo} model")
    model=GridSearchCV(pipeline,grid[algo],n_jobs=-1,cv=10)
    model.fit(x_train,y_train)
    fit_model[algo]=model
```

# Evaluation Peformance on test partition

```python
# fit_model.items()
```

```python
from sklearn.metrics import accuracy_score,recall_score,precision_score
```

```python
for algo,model in fit_model.items():
    yhat=model.predict(x_test)
    acc_score=accuracy_score(y_test,yhat)
    pres_score=precision_score(y_test,yhat)
    rec_score=recall_score(y_test,yhat)
    print(f"accuracy of {algo} is-{acc_score}  and recall-{rec_score} and precision-{pres_score}")
    
```
